```python
def summary(dataset, column):
    print('mean is ', data[column].mean())
    print('median is: ', data[column].median())
    print('mode is: ', data[column].mode()[0])
    print('variance is: ', data[column].var())
    print('standard deviation is: ', data[column].std())
    print('range is: (', data[column].min(), ',', data[column].max(),')')
```

```python
columns = ['Points','Score','Weigh']
```

```python
matrix = [[data[x].mean() for x in columns]]
matrix.append([data[x].median() for x in columns])
matrix.append([data[x].mode()[0] for x in columns])
matrix.append([data[x].var() for x in columns])
matrix.append([data[x].std() for x in columns])
matrix.append(['('+str(data[x].min())+','+str(data[x].max())+')' for x in columns])
```

```python
q7 = pd.DataFrame(matrix, columns=['Points','Score','Weigh'], index=['mean','median','mode','variance','std','range'])
```
